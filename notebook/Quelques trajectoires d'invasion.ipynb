Voici une exploration de trajectoire d'invasion et de mesures de la fitness d'invasion

```python
from __future__ import division
from pimad.models.toycontinuous import ToyContinuous
import matplotlib.pyplot as plt
import pimad.export.draw as draw
import numpy as np
plt.rc("figure",figsize=(15,15))
%matplotlib inline
```

```python
def trajectoire(**kargs):
    # Default parameters:
    param = {"n":100,"T":1000,"ip":0,"m":0.5,"r":0.5,"b":20,"c":1,"g":1000,"mu":1e-5,"replicas":5}
    
    # Overiding default parameters
    for k,v in kargs.items():
        param[k] = v

    all_traj = []
    for n in range(param["replicas"]):
        m=ToyContinuous(param)
        traj = np.zeros(m.p["g"])

        for i in range(m.p["g"]):
            m.step()
            traj[i] = m.population.mutants.sum()
        all_traj.append(traj)
    
    all_traj = np.array(all_traj)
    mean = all_traj.mean(0)
    std = all_traj.std(0)
    return mean,std
```

```python
neutral = trajectoire()   
social = trajectoire(m=0.51)
asocial = trajectoire(m=0.49)
```

```python
plt.plot(range(len(neutral[0])),neutral[0],label="neutral (0.50)",color="blue")
plt.fill_between(range(len(neutral[0])),neutral[0]-neutral[1],neutral[0]+neutral[1],color="blue",alpha=0.1)

plt.plot(range(len(social[0])),social[0],label="social (0.51)",color="green")
plt.fill_between(range(len(social[0])),social[0]-social[1],social[0]+social[1],color="green",alpha=0.1)

plt.plot(range(len(asocial[0])),asocial[0],label="asocial (0.49)",color="red")
plt.fill_between(range(len(asocial[0])),asocial[0]-asocial[1],asocial[0]+asocial[1],color="red",alpha=0.1)

plt.legend()
plt.xlabel("generation")
plt.ylabel("number of mutant individuals")
plt.show()
```

Using the estimation of the invasion fitness used in the report: 

$$f = \log_{10}\left(\frac{M(10)}{M(0)}\right) $$

```python
print "Social:", np.log10(social[0][10]/social[0][0])
print "Neutral:", np.log10(neutral[0][10]/neutral[0][0])
print "Asocial:", np.log10(asocial[0][10]/asocial[0][0])
```
